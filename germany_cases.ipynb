```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## manually collected data including direct early announcments from lab until Mar. 16
rkim = pd.read_csv('rki_manual.csv', index_col=0, parse_dates=[0])

## electronically submitted data (lower numbers) starting Mar. 10
rkie = pd.read_csv('rki_electronic.csv', index_col=0, parse_dates=[0])
```

```python
rkie.tail()
```

```python
# choose a random list of federal states to plot
plot_de = [
    'Hamburg',
    'Bremen',
    'Berlin',
    'Bayern',
    'Nordrhein-Westfalen'
]
```

```python
plt.close(1)
fig1, ax1 = plt.subplots(ncols=2, figsize=(10,4), sharex=True, num=1)
rkim[plot_de].plot(ax=ax1[0], logy=True) 
rkim[plot_de].diff().plot(ax=ax1[1], logy=False)

ax1[0].set_ylabel('Confirmed')
ax1[1].set_ylabel('New Infections')

fig1.tight_layout()
```

```python
plt.close(2)
fig2, ax2 = plt.subplots(ncols=2, figsize=(10,4), sharex=True, num=2)
rkie[plot_de].plot(ax=ax2[0], logy=True) 
rkie[plot_de].diff().plot(ax=ax2[1], logy=False)

ax2[0].set_ylabel('Confirmed')
ax2[1].set_ylabel('New Infections')

fig2.tight_layout()
```

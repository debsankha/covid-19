```python
url = "https://opendata.arcgis.com/datasets/dd4580c810204019a7b8eb3e0b329dd6_0.csv"

df = pd.read_csv(url)
df
```

```python
df['Meldedatum'] = pd.to_datetime(df["Meldedatum"])
```

```python
nie = df[df['Bundesland']=='Niedersachsen'][['Meldedatum', 'AnzahlFall']]
nie
```

```python
nie['date'] = nie['Meldedatum'].apply(lambda x:x.date())
nie
```

```python
case_ts = nie[['date', 'AnzahlFall']].groupby('date').agg('sum')
case_ts['cumulative'] = case_ts['AnzahlFall'].cumsum()
case_ts
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

case_ts['cumulative'].plot(ax=ax1)

thres = 50
thres_crossing_date = case_ts[case_ts['cumulative'] > 50].iloc[0].name
case_ts['day_since_thres'] = case_ts.index - thres_crossing_date

sns.lineplot(x='day_since_thres', y='cumulative', estimator=None, data=case_ts)
day_min = 0
day_max = case_ts['day_since_thres'].max()/pd.Timedelta(days=1)

ax2.set_xlim(0, day_max*24*3600*1000)

xtick_days = np.arange(0, day_max, 5)
xticks = [day*24*3600*1000 for day in xtick_days]
xticklabels = [str(d) for d in xtick_days]

ax2.set_xticks(xticks)
ax2.set_xticklabels(xticklabels)
```

# Cases in each state

```python


fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18,7))

day_max = 0

for bundesland in df['Bundesland'].unique():
    df_state = df[df['Bundesland']==bundesland][['Meldedatum', 'AnzahlFall']]
    df_state['date'] = df_state['Meldedatum'].apply(lambda x:x.date())
    case_ts = df_state[['date', 'AnzahlFall']].groupby('date').agg('sum')
    case_ts['cumulative'] = case_ts['AnzahlFall'].cumsum()

    case_ts['cumulative'].plot(ax=ax1, label=bundesland[:4])
    ax1.tick_params(axis='x', labelrotation=45)


    thres = 50
    if len(case_ts[case_ts['cumulative'] > 50]) == 0:
        continue
    thres_crossing_date = case_ts[case_ts['cumulative'] > 50].iloc[0].name
    case_ts['day_since_thres'] = (case_ts.index - thres_crossing_date)/pd.Timedelta(days=1)

    sns.lineplot(x='day_since_thres', y='cumulative', estimator=None, data=case_ts, ax=ax2, label=bundesland[:4])
    
ax2.set_xlim(0, ax2.get_xlim()[1])
ax1.legend()
ax2.legend()

ax2.set_xlabel("days since 50 cases")
ax2.set_ylabel("cases")

ax2.set_yscale('log')
ax1.set_yscale('log')
```

```python
import pandas as pd
```

```python
tables = pd.read_html("https://de.wikipedia.org/wiki/Liste_der_deutschen_Bundesl%C3%A4nder_nach_Bev%C3%B6lkerung", thousands='.')
tables[0]
```

```python
population_states = tables[0][['Bundesland', '2018']]
population_states.head()
```

```python
population_states.loc[2, 'Bundesland'] = 'Berlin'
population_states.loc[16, 'Bundesland'] = 'Germany'
```

```python
population_states
```

```python
day_max = 0

with sns.color_palette("husl"):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18,7))
    
    case_thres = 50

    for bundesland in list(df['Bundesland'].unique()) + ['Germany']:
        # get population
        try:
            pop = population_states[population_states['Bundesland']==bundesland].values[0,1]
        except:
            print(f"Population for {bundesland} not found")
            continue
                    
        if bundesland == 'Germany':
            label='Germany'
            df_state = df[['Meldedatum', 'AnzahlFall']]
        else:
            df_state = df[df['Bundesland']==bundesland][['Meldedatum', 'AnzahlFall']]
            label = bundesland[:4]
            
        df_state['date'] = df_state['Meldedatum'].apply(lambda x:x.date())
        df_state.sort_values(by='date')

        case_ts = df_state[['date', 'AnzahlFall']].groupby('date').agg('sum')
        case_ts['cumulative'] = case_ts['AnzahlFall'].cumsum()
        case_ts['cumulative_normed'] = case_ts['cumulative']/pop*100000

        case_ts['cumulative_normed'].plot(ax=ax1, label=label)
        ax1.tick_params(axis='x', labelrotation=45)


        if len(case_ts[case_ts['cumulative'] > case_thres]) == 0:
            continue
        thres_crossing_date = case_ts[case_ts['cumulative'] > case_thres].iloc[0].name
        case_ts['day_since_thres'] = (case_ts.index - thres_crossing_date)/pd.Timedelta(days=1)

        sns.lineplot(x='day_since_thres', y='cumulative_normed', estimator=None, data=case_ts, ax=ax2, label=label, markers=True, dashes=True)
        
    ax2.set_xlim(0, ax2.get_xlim()[1])
    ax1.legend(fontsize=15, ncol=2)
    ax2.legend(fontsize=15, ncol=2)

    ax2.set_xlabel(f"days since {case_thres} cases reported")
    ax2.set_ylabel("cases/100000 inhabitants")

    ax2.set_yscale('log')
    ax1.set_yscale('log')

fig.tight_layout()

fig.savefig("bundeslaender.pdf")
```

# Deaths

```python
day_max = 0

with sns.color_palette("husl"):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18,7))
    
    for bundesland in list(df['Bundesland'].unique()) + ['Germany']:
        # get population
        try:
            pop = population_states[population_states['Bundesland']==bundesland].values[0,1]
        except:
            print(f"Population for {bundesland} not found")
            continue
                    
        if bundesland == 'Germany':
            label='Germany'
            df_state = df[['Meldedatum', 'AnzahlTodesfall']]
        else:
            df_state = df[df['Bundesland']==bundesland][['Meldedatum', 'AnzahlTodesfall']]
            label = bundesland[:4]
            
        df_state['date'] = df_state['Meldedatum'].apply(lambda x:x.date())
        df_state.sort_values(by='date')
        case_ts = df_state[['date', 'AnzahlTodesfall']].groupby('date').agg('sum')
        case_ts['cumulative'] = case_ts['AnzahlTodesfall'].cumsum()
        case_ts['cumulative_normed'] = case_ts['cumulative']/pop*100000

        case_ts['cumulative_normed'].plot(ax=ax1, label=label)
        ax1.tick_params(axis='x', labelrotation=45)


        if len(case_ts[case_ts['cumulative'] > 0]) == 0:
            print(f"No death in {bundesland}")
            continue
        thres_crossing_date = case_ts[case_ts['cumulative'] > 0].iloc[0].name
        case_ts['day_since_thres'] = (case_ts.index - thres_crossing_date)/pd.Timedelta(days=1)

        sns.lineplot(x='day_since_thres', y='cumulative_normed', estimator=None, data=case_ts, ax=ax2, label=label, markers=True, dashes=True)
        
    ax2.set_xlim(0, ax2.get_xlim()[1])
    ax1.legend(fontsize=15, ncol=2)
    ax2.legend(fontsize=15, ncol=2)

    ax2.set_xlabel(f"days since first death")
    ax2.set_ylabel("deaths/100000 inhabitants")

    ax2.set_yscale('log')
    ax1.set_yscale('log')

fig.tight_layout()

#fig.savefig("bundeslaender.pdf")
```

```python
day_max = 0

with sns.color_palette("husl"):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18,7))
    
    case_thres = 50

    for bundesland in list(df['Bundesland'].unique()) + ['Germany']:
        # get population
        try:
            pop = population_states[population_states['Bundesland']==bundesland].values[0,1]
        except:
            print(f"Population for {bundesland} not found")
            continue
                    
        if bundesland == 'Germany':
            label='Germany'
            df_state = df[['Meldedatum', 'AnzahlFall']]
        else:
            df_state = df[df['Bundesland']==bundesland][['Meldedatum', 'AnzahlFall']]
            label = bundesland[:4]
            
        df_state['date'] = df_state['Meldedatum'].apply(lambda x:x.date())
        df_state.sort_values(by='date')

        case_ts = df_state[['date', 'AnzahlFall']].groupby('date').agg('sum')
        case_ts['cumulative'] = case_ts['AnzahlFall'].cumsum()
        case_ts['cumulative_normed'] = case_ts['cumulative']/pop*100000

        case_ts['cumulative_normed'].plot(ax=ax1, label=label)
        ax1.tick_params(axis='x', labelrotation=45)


        if len(case_ts[case_ts['cumulative'] > case_thres]) == 0:
            continue
        thres_crossing_date = case_ts[case_ts['cumulative'] > case_thres].iloc[0].name
        case_ts['day_since_thres'] = (case_ts.index - thres_crossing_date)/pd.Timedelta(days=1)

        sns.lineplot(x='day_since_thres', y='cumulative_normed', estimator=None, data=case_ts, ax=ax2, label=label, markers=True, dashes=True)
        
    ax2.set_xlim(0, ax2.get_xlim()[1])
    ax1.legend(fontsize=15, ncol=2)
    ax2.legend(fontsize=15, ncol=2)

    ax2.set_xlabel(f"days since {case_thres} cases reported")
    ax2.set_ylabel("cases/100000 inhabitants")

    ax2.set_yscale('log')
    ax1.set_yscale('log')

fig.tight_layout()

fig.savefig("bundeslaender.pdf")
```

```python
day_max = 0

with sns.color_palette("husl"):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18,7))
    
    for bundesland in list(df['Bundesland'].unique()) + ['Germany']:
        # get population
        try:
            pop = population_states[population_states['Bundesland']==bundesland].values[0,1]
        except:
            print(f"Population for {bundesland} not found")
            continue
                    
        if bundesland == 'Germany':
            label='Germany'
            df_state = df[['Meldedatum', 'AnzahlTodesfall']]
        else:
            df_state = df[df['Bundesland']==bundesland][['Meldedatum', 'AnzahlTodesfall']]
            label = bundesland[:4]
            
        df_state['date'] = df_state['Meldedatum'].apply(lambda x:x.date())
        df_state.sort_values(by='date')
        case_ts = df_state[['date', 'AnzahlTodesfall']].groupby('date').agg('sum')
        case_ts['cumulative'] = case_ts['AnzahlTodesfall'].cumsum()
        case_ts['cumulative_normed'] = case_ts['cumulative']/pop*100000

        case_ts['cumulative_normed'].plot(ax=ax1, label=label)
        ax1.tick_params(axis='x', labelrotation=45)


        if len(case_ts[case_ts['cumulative'] > 0]) == 0:
            print(f"No death in {bundesland}")
            continue
        thres_crossing_date = case_ts[case_ts['cumulative'] > 0].iloc[0].name
        case_ts['day_since_thres'] = (case_ts.index - thres_crossing_date)/pd.Timedelta(days=1)

        sns.lineplot(x='day_since_thres', y='cumulative_normed', estimator=None, data=case_ts, ax=ax2, label=label, markers=True, dashes=True)
        
    ax2.set_xlim(0, ax2.get_xlim()[1])
    ax1.legend(fontsize=15, ncol=2)
    ax2.legend(fontsize=15, ncol=2)

    ax2.set_xlabel(f"days since first death")
    ax2.set_ylabel("deaths/100000 inhabitants")

    ax2.set_yscale('log')
    ax1.set_yscale('log')
    ax1.xaxis.set_tick_params(rotation=45)
fig.tight_layout()

#fig.savefig("bundeslaender.pdf")
```

## get all the states in a single DF

```python
df_state = df[df['Bundesland']==bundesland][['Meldedatum', 'AnzahlFall']]
            
df_state['date'] = df_state['Meldedatum'].apply(lambda x:x.date())
df_state.sort_values(by='date')
```

```python
narrow_df = df[['Bundesland', 'AnzahlFall', 'AnzahlTodesfall', 'Meldedatum']]
narrow_df['Meldedatum'] = narrow_df['Meldedatum'].apply(lambda x:x.date())
narrow_df
```

```python
narrow_df.sort_values(by='Meldedatum', inplace=True)
cumsum_dfs = []

for bundesland in list(narrow_df['Bundesland'].unique())+['Germany']:
    if bundesland == 'Germany':
        sel = narrow_df
    else:
        sel = narrow_df[narrow_df['Bundesland']==bundesland]
    sel = sel[['Meldedatum','AnzahlFall', 'AnzahlTodesfall']].groupby('Meldedatum').agg('sum')
    sel = sel.sort_values(by='Meldedatum')
    sel['cumulative_cases'] = sel['AnzahlFall'].cumsum()
    sel['cumulative_deaths'] = sel['AnzahlTodesfall'].cumsum()
    
    if len(sel[sel['cumulative_cases'] > 50]) > 0:
        first_case_date = sel[sel['cumulative_cases'] > 50].iloc[0].name
        sel['days_since_50_cases'] = (sel.index - first_case_date)/pd.Timedelta(days=1)
    else:
        sel['days_since_50_cases'] = None
    
    if len(sel[sel['cumulative_deaths'] > 0]) > 0:
        first_death_date = sel[sel['cumulative_deaths'] > 0].iloc[0].name
        sel['days_since_first_death'] = (sel.index - first_death_date)/pd.Timedelta(days=1)
    else:
        sel['days_since_first_death'] = None
    sel['Bundesland'] = bundesland
    
    cumsum_dfs.append(sel.reset_index())

all_cum_df = pd.concat(cumsum_dfs)
all_cum_df
```

```python
german_state_abbrs = {
'Baden-Württemberg': 'BW',
'Bayern': 'BY',
'Berlin': 'BE',
'Brandenburg': 'BB',
'Bremen': 'HB',
'Hamburg': 'HH',
'Hessen': 'HE',
'Niedersachsen': 'NI',
'Mecklenburg-Vorpommern': 'MV',
'Nordrhein-Westfalen': 'NW',
'Rheinland-Pfalz': 'RP',
'Saarland': 'SL',
'Sachsen': 'SN',
'Sachsen-Anhalt': 'ST',
'Schleswig-Holstein': 'SH',
'Thüringen': 'TH',
}
```

```python
all_cum_df['state'] = all_cum_df['Bundesland'].apply(lambda x:german_state_abbrs[x] if x!='Germany' else 'Germany')
all_cum_df['date'] = all_cum_df['Meldedatum'].apply(lambda x:pd.Timestamp(x))

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18,6))

sns.lineplot(data=all_cum_df, x='Meldedatum', y='cumulative_cases_normed', hue='state', ax=ax1, ci=None)
sns.lineplot(data=all_cum_df, x='days_since_50_cases', y='cumulative_cases_normed', hue='state', ax=ax2, ci=None)

#ax1.set_xlim(0, ax1.get_xlim()[1])
ax2.set_xlim(0, ax2.get_xlim()[1])
ax1.legend(fontsize=10, ncol=2)
ax2.legend(fontsize=10, ncol=2)

ax1.set_ylabel("cases/100000 inhabitants")
ax2.set_xlabel(f"days since 50 cases")
ax2.set_ylabel("cases/100000 inhabitants")

ax1.xaxis.set_tick_params(rotation=45)

ax2.set_yscale('log')
ax1.set_yscale('log')

fig.tight_layout()
fig.savefig("bundeslaender_faelle.pdf")
```
